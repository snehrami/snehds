```python
import numpy as np
import pandas as pd

df=pd.read_csv("/content/sample_data/data.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
for coll in df.columns:
    if df[coll].dtype == 'object':
        df[coll] = df[coll].fillna(df[coll].mode().iloc[0] if not df[coll].mode().empty else "Unknown")
    else:
        df[coll] = df[coll].fillna(df[coll].median() if not df[coll].dropna().empty else 0)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap (Numeric Columns Only)")
plt.show()
```

```python
sns.pairplot(df[df.columns[:5]])
plt.show()
```

```python
sns.histplot(df['MSRP'], bins=30, kde=True)
plt.title("Distribution of Car Prices")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])
print(numeric_df.corr()['MSRP'].sort_values(ascending=False))
```

```python
df.to_csv("cleaned_car_data.csv", index=False)
```
